```python
import pandas as pd
import numpy as np 
import os 
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
X= iris.data
y= iris.target
```

```python
df = pd.DataFrame(X,columns= iris.feature_names)
df['species'] = y
df
```

```python
import matplotlib.pyplot as plt 

fig, axes = plt.subplots(1, 4, figsize=(14, 4))
axes[0].hist(df['petal length (cm)'],bins=20,color = 'green')
axes[0].set_title('petal length')
axes[1].hist(df['petal width (cm)'],bins=20,color = 'red')
axes[1].set_title('petal width')
axes[2].hist(df['sepal length (cm)'],bins=20,color = 'blue')
axes[2].set_title('sepal length')
axes[3].hist(df['sepal width (cm)'],bins=20,color = 'yellow')
axes[3].set_title('sepal width')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns 

fig, axes = plt.subplots(4, 1, figsize=(18, 14))
sns.boxplot(x = df['sepal length (cm)'],ax = axes[0])
sns.boxplot(x = df['petal length (cm)'],ax = axes[1])
sns.boxplot(x = df['petal width (cm)'],ax = axes[2])
sns.boxplot(x = df['sepal width (cm)'],ax = axes[3])
```

```python
sns.countplot(x=df['species'])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,6))
sns.scatterplot(x= df['petal length (cm)'],y=df['petal width (cm)'],hue =df['species'],ax=axes[0],palette='deep')

sns.scatterplot(x= df['sepal length (cm)'],y=df['sepal width (cm)'],hue =df['species'],ax=axes[1],palette='Set2')
```

```python
sns.pairplot(df, hue='species',palette='deep')
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```
